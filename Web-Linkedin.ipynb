```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from time import sleep
import csv



print('-Finishing import library package')
```

```python
# Step 1 : Login to Linkedin 

# Open Chrome and login Linkedin login site
driver = webdriver.Chrome()
url = 'https://www.linkedin.com/login'
driver.get(url)
driver.maximize_window()
print('Finished initialize a driver')

sleep(2)

#Import user name and password
credential = open('login_credential.txt')
line = credential.readlines()
username =line[0]
password = line[1]

print('Finished importing the login cendential')

#Key in username
email_field = driver.find_element("id",'username')
email_field.send_keys(username)
print('Finished email')
sleep(2)

#Key in password
password_field = driver.find_element("name","session_password")
password_field.send_keys(password)
print('Finished password')
sleep(3)

#Click login button
login_field = driver.find_element('xpath','//*[@id="organic-div"]/form/div[3]/button')
login_field.click()
print('Finshed Login')
sleep(2)
```

```python
# Step 2 : Searching for Software engineer people 

#Locate the search bar element
searching_field = driver.find_element('xpath','//*[@id="global-nav-typeahead"]/input')

#Input the search keyword to search bar
search_query = input('What keyword do you to search?')
searching_field.send_keys(search_query)
sleep(2)
#Search
searching_field.send_keys(Keys.RETURN)

sleep(3)

#Click to people button 

People_button = driver.find_element('xpath', '//button[text()="People"]')
People_button.click()
print('Finshed searching')
sleep(1)
```

```python
# Step 3: Scarping data with BeautifulSoup 
def GetURL():
    page_source = BeautifulSoup(driver.page_source)
    profiles = page_source.select('a.app-aware-link.scale-down')
    all_profile_URL = []
    for profile in profiles:
        profile_link = profile.get('href')
        html = profile_link.split('?miniProfileUrn')[0]
        all_profile_URL.append(html)

    return all_profile_URL

input_page = int(input('how many pages do you want to scrape'))
URLs_all_page = []
for page in range(input_page):
    URLs_one_page = GetURL()
    sleep(1)
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    sleep(1)
    next_button = driver.find_element(By.CLASS_NAME, 'artdeco-pagination__button--next')
    next_button.click()
    URLs_all_page = URLs_all_page + URLs_one_page
    sleep(1)

print(URLs_all_page)

print('Finished Task 3')
```

```python
#Task 4 : Scrape data of 1 profile and write the data to a CSV file

with open('output.csv','w',newline= '',encoding='utf-8') as file_output:
    headers = ['Name','Location', 'Job Title', 'URL']
    writer = csv.DictWriter(file_output, delimiter =',', lineterminator = '\n', fieldnames = headers)
    writer.writeheader()
    
    for linkedin_URL in URLs_all_page:
        driver.get(linkedin_URL)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source,"html.parser")

        #Name
        info_div = soup.find('div' ,class_= "mt2")
        info_loc = info_div.find_all('div',class_ = "pv-text-details__left-panel")
        name = info_loc[0].find('h1').get_text().strip()
        print('Profile name is :' ,name)

        #Location

        info_Location = info_div.select('div.pv-text-details__left-panel.mt2')
        location = info_Location[0].find('span').get_text().strip()
        print('Profile location is :' ,location)

        #Title

        title = info_div.find('div',class_ = "text-body-medium break-words").get_text().strip()
        print('Profile Title is :' ,title)
        print('\n')
        writer.writerow({headers[0]:name, headers[1]:location, headers[2]:title, headers[3]:linkedin_URL},)

print('Finished Scraping')

```
